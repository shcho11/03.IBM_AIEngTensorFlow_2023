### Split & Normalize Dataset

```python
from sklearn.model_selection import train_test_split

concrete_data_columns = concrete_data.columns
x = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] 
y = concrete_data['Strength']
```

```python
x.head()
```

```python
y.head()
```

```python
x_norm = (x - x.mean()) / x.std()
x.head()
```

```python
n_cols = x_norm.shape[1]
n_cols
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_norm, y, test_size = 0.3, random_state = 42)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# define regression model
def reg_model1():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#tf.compat.v1.get_default_graph 
```

```python
model = reg_model1()
```

```python
# fit the model
model.fit(x_train, y_train, validation_split=0.3, epochs=100, verbose=2)
```

```python
# Evaluate the model on the test data using 'evaluate'
print("Evaluate on test data")
results = model.evaluate(x_test, y_test)
print("test loss : ", results)

y_pred = model.predict(x_test)
print("prediction shape:", y_pred.shape)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
sd_mse = np.std(mse)
print(mse, sd_mse)
```

```python
times = 50 # Repeat times unchanged 
epochs = 100 # epochs changed from 50 to 100
mse_list = []

for i in range(1, times+1) : 
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
  model.fit(x_train, y_train, validation_split=0.3, epochs=epochs, verbose=0)
  results = model.evaluate(x_test, y_test, verbose=0)
  print("MSE {} : {}".format(i,results))
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test, y_pred)
  mse_list.append(mse)

mse_list = np.array(mse_list)
mean_mse_list = np.mean(mse_list)
sd_mse_list = np.std(mse_list)

print("\n")
print("Results(Question C : With Normalization and 100 epochs) :", "\n")
print("Mean of 50 MSEs: {}".format(mean_mse_list))
print("Standard deviation of 50 MSEs : {}".format(sd_mse_list))
```
